### Credit Card Fraud Detection
This project aims to visualize and create a reliable model to predict credit card fraud. The dataset is aviable on Kaggle and was originally colected and analysed during a research collaboration of Wordline and the Machine Learning Group of the Université Libre de Bruxelles. 

The dataset contains real transactions made in 2013 by European cardholders during two days, the individuals were anonymized.

This will be a problem... the class data, which is the fraud data, is heavily imbalanced. I will try to deal with it by applying under and oversampling combined. The method applied will be Smote for oversampling and Tomek for undersamplying. To compare, I will create another model of over and undersampling using Smote and ENN.

Before, I will need to reduce the number of feature.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif


X = data.drop(columns='Class', axis=1)
y = data['Class']

best_features = SelectKBest(f_classif, k = 20)
best_features.fit(X, y)
best_Xv2 = best_features.transform(X)
print(best_Xv2.shape)
```

```python
from imblearn.combine import SMOTETomek
from sklearn.model_selection import train_test_split

TrainX, TestX, Trainy, Testy = train_test_split(best_Xv2, y, test_size=0.3, random_state=1405,  stratify = y)

smt = SMOTETomek(random_state=1405)
TrainX_smt, Trainy_smt = smt.fit_resample(TrainX, Trainy)

print(Trainy_smt.value_counts())
print(TrainX_smt.shape)
print(TrainX.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
knn = KNeighborsClassifier(metric='euclidean')
knn.fit(TrainX_smt, Trainy_smt)
predict_knn_smt = knn.predict(TestX)
```

```python
print(np.median(TrainX_smt))
```

```python
bnb = BernoulliNB(binarize=np.median(TrainX_smt))
bnb.fit(TrainX_smt, Trainy_smt)
predict_bnb_smt = bnb.predict(TestX)
```

```python
dtc = DecisionTreeClassifier(criterion='entropy', random_state=1405)
dtc.fit(TrainX_smt, Trainy_smt)
predict_dtc_smt = dtc.predict(TestX)
```

```python
clf = LogisticRegression(random_state=1405)
clf.fit(TrainX_smt, Trainy_smt)
predict_clf_smt = dtc.predict(TestX)
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
recall_knn_smt = recall_score(Testy, predict_knn_smt)
recall_bnb_smt = recall_score(Testy, predict_bnb_smt)
recall_dtc_smt = recall_score(Testy, predict_dtc_smt)
recall_clf_smt = recall_score(Testy, predict_clf_smt)
```

```python
precision_knn_smt = precision_score(Testy, predict_knn_smt)
precision_bnb_smt = precision_score(Testy, predict_bnb_smt)
precision_dtc_smt = precision_score(Testy, predict_dtc_smt)
precision_clf_smt = precision_score(Testy, predict_clf_smt)
```

```python
Results = {'precision':[precision_knn_smt, precision_bnb_smt, precision_dtc_smt, precision_clf_smt],
           'recall':[recall_knn_smt, recall_bnb_smt, recall_dtc_smt, recall_clf_smt]}

Results = pd.DataFrame(Results, index=['KNearestNeighbors', 'BernoulliNB', 'DecisionTreeClassifier', 'Logit'])

Results
```
